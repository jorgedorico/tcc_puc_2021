## Processamento/Tratamentos de Dados

```python
# Importando Bibliotecas
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Carregando a base
conf_df = pd.read_csv('tcc_conformidade.csv', dtype={ 'cnae_classe': 'string'}, sep=';', decimal =',')
conf_df.head()
```

```python
#Verificando colunas e tipos atribuídos
print('Base de Dados: {} linhas e {} colunas'.format(len(conf_df), len(conf_df.columns)),'\n\n')
conf_df.info()
```

```python
# Extraindo código e descrição da coluna setor
setor = conf_df.setor

setor_cod=[]
setor_desc=[]

for i in range(0,len(setor)):
    setor_partes=setor[i].split(' ',2)
    setor_cod.append(setor_partes[0][0:4]+setor_partes[0][4])
    setor_desc.append(setor_partes[2])
    
setor_df=pd.DataFrame(list(zip(setor_cod, setor_desc)), columns = ['setor_cod', 'setor_desc'])
setor_df.head()
```

```python
# Concatenando Dataframes
conf_df = pd.concat([setor_df, conf_df], axis = 1)
conf_df.head()
```

```python
# Removendo Colunas desnecessárias
conf_df = conf_df.drop(columns=['setor','setor_desc','grande_empresa'])
print('Base de Dados: {} linhas e {} colunas'.format(len(conf_df), len(conf_df.columns)),'\n\n')
conf_df.info()
```

```python
# Ajustando Tipo da Coluna, conforme base original
tipo_coluna = {'setor_cod': 'string',
               'situacao_cadastral': 'string',
               'natureza_juridica': 'string',
               'impacto': 'string'}
conf_df = conf_df.astype(tipo_coluna)
conf_df.dtypes
```

```python
# Verificando as colunas de Dimensão e Conformidade (valores entre 0 e 1)
obrigacoes_acessorias_0=conf_df.loc[conf_df['obrigacoes_acessorias'] < 0].empty  
obrigacoes_acessorias_1=conf_df.loc[conf_df['obrigacoes_acessorias'] > 1].empty  
obrigacao_principal_0=conf_df.loc[conf_df['obrigacao_principal'] < 0].empty  
obrigacao_principal_1=conf_df.loc[conf_df['obrigacao_principal'] > 1].empty  
atuacao_fiscal_0=conf_df.loc[conf_df['atuacao_fiscal'] < 0].empty  
atuacao_fiscal_1=conf_df.loc[conf_df['atuacao_fiscal'] > 1].empty  
analise_comparativa_0=conf_df.loc[conf_df['analise_comparativa'] < 0].empty  
analise_comparativa_1=conf_df.loc[conf_df['analise_comparativa'] > 1].empty  
consistencia_0=conf_df.loc[conf_df['consistencia'] < 0].empty  
consistencia_1=conf_df.loc[conf_df['consistencia'] > 1].empty  
comercio_exterior_0=conf_df.loc[conf_df['comercio_exterior'] < 0].empty  
comercio_exterior_1=conf_df.loc[conf_df['comercio_exterior'] > 1].empty  
conformidade_0=conf_df.loc[conf_df['conformidade'] < 0].empty   
conformidade_1=conf_df.loc[conf_df['conformidade'] > 1].empty   

print('Obrigações Acessórias 0 - {:}'.format(obrigacoes_acessorias_0))
print('Obrigações Acessórias 1 - {:}'.format(obrigacoes_acessorias_1))
print('Obrigação Principal 0 - {:}'.format(obrigacao_principal_0))
print('Obrigação Principal 1 - {:}'.format(obrigacao_principal_1))
print('Atuação Fiscal 0 - {:}'.format(atuacao_fiscal_0))
print('Atuação Fiscal 1 - {:}'.format(atuacao_fiscal_1))
print('Análise Comparativa 0 - {:}'.format(analise_comparativa_0))
print('Análise Comparativa 1 - {:}'.format(analise_comparativa_1))
print('Consistência 0 - {:}'.format(consistencia_0))
print('Consistência 1 - {:}'.format(consistencia_1))
print('Comércio Exterior 0 - {:}'.format(comercio_exterior_0))
print('Comércio Exterior 1 - {:}'.format(comercio_exterior_1))
print('Conformidade 0 - {:}'.format(conformidade_0))
print('Conformidade 1 - {:}'.format(conformidade_1))
```
